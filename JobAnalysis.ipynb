## 資料清理

### 丟掉欄位

```python
df = df.drop(['hrName','email','phone','addressDetail','manageResp', 'businessTrip','hireType','workPeriod','language',
              'vacationPolicy','startWorkingDay','acceptRole','jobType','workType', 'welfare_tag', 'welfare', 'legalTag',
             'workExp', 'edu'], axis=1)
```

### 因為有很多掛著好看的職缺，我們在這裡將時間限定在一個月內
9/20抓的資料，限定日期大於8/20

```python
df['appearDate'] = df['appearDate'].apply(lambda x: datetime.datetime.strptime(x, '%Y/%m/%d'))
df = df.loc[df['appearDate']>='2020-08-21',:]
df.info()
```

```python
df = df.drop(['appearDate'], axis=1)
```

### 只要月薪的資料
- 待遇面議表示月薪4萬以上
- 先丟掉日薪、時薪的工作

```python
df.groupby('salaryType').head(1)
```

```python
df = df.loc[df['salaryType'].apply(lambda x: x in [10, 50, 60]),:]
df.shape
```

## 特徵萃取

### 新增index

```python
df= df.reset_index()
df.head()
```

### 月薪重分類
- 年薪除14計算月薪
- 面議為4萬以上

```python
# 轉換成月薪
df['nsalaryMin'] = np.select(condlist = [df['salaryType'] == 60, # 年薪
                                         df['salaryType'] == 10, # 面議
                                         df['salaryType'] == 50], # 月薪
                             choicelist = [df['salaryMin']/14,
                                           40000,
                                           df['salaryMin']])
```

- 透過觀察的方式發現月薪超過 30萬以上的資料很可能是寫錯資料，在這裡將資料濾除掉

```python
print(df.shape)
df = df.loc[df['nsalaryMin']<=300000, :]
print(df.shape)
```

```python
df['nsalaryMin'].min()
```

```python
# 月薪重分類
df['nsalaryMin'] = np.select(condlist = [df['nsalaryMin'] < 25000,
                                         df['nsalaryMin'] < 30000,
                                         df['nsalaryMin'] < 35000,
                                         df['nsalaryMin'] < 40000,
                                         df['nsalaryMin'] >= 40000],
                             choicelist = [r'25k-', r'25-30k', r'30-35k', r'35-40k', r'40k+'])
```

```python
df = df.drop(['salary', 'salaryMin', 'salaryMax', 'salaryType'], axis =1)
```

### 職務內容
許多對職缺的要求寫在其他裡面，在這裡將資料串接起來

```python
df['jobDetail'] = df['jobDetail'] + r'\r\n'+ df['other']
```

```python
df = df.drop(['other'], axis =1)
```
