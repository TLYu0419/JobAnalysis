### 月薪重分類
- 年薪除14計算月薪
- 面議為4萬以上
- 通常新鮮人沒有籌碼跟公司談薪資，所以這裡取最低薪資

```python
# 轉換成月薪
df['nsalaryMin'] = np.select(condlist = [df['salaryType'] == 60, # 年薪
                                         df['salaryType'] == 10, # 面議
                                         df['salaryType'] == 50], # 月薪
                             choicelist = [df['salaryMin']/14,
                                           40000,
                                           df['salaryMin']])
```

- 透過觀察的方式發現月薪超過 30萬以上的資料很可能是寫錯資料，在這裡將資料濾除掉

```python
print(df.shape)
df = df.loc[df['nsalaryMin']<=300000, :]
print(df.shape)
```

```python
df['nsalaryMin'].min()
```

```python
# 月薪重分類
df['nsalaryMin'] = np.select(condlist = [df['nsalaryMin'] < 25000,
                                         df['nsalaryMin'] < 30000,
                                         df['nsalaryMin'] < 35000,
                                         df['nsalaryMin'] < 40000,
                                         df['nsalaryMin'] >= 40000],
                             choicelist = [r'25k-', r'25-30k', r'30-35k', r'35-40k', r'40k+'])
```

### 斷詞

```python
import spacy
from spacy.matcher import Matcher
nlp = spacy.load('zh_core_web_lg')
```

```python
doc2 = nlp(Soc['jobDetail'][2])
for token in doc2:
    print(token.text, token.pos_, token.is_stop)
```

```python
def sent2seg(sent):
    sent2 = nlp(sent)
    return ' '.join([ token.text for token in sent2 if token.pos_ in ['NOUN', 'VERB'] and token.is_stop == False ])
```

```python
sent2seg(Soc['jobDetail'][2])
```

```python
Soc['seg'] = Soc['jobDetail'].apply(lambda x: sent2seg(x))
Soc.head(3)
```

### 文字雲
- 這邊需要說明一下文字編碼的問題

```python
from wordcloud import WordCloud, ImageColorGenerator
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
doc = [' '.join(Soc['seg'])]
# Create a Vectorizer Object 
vectorizer = CountVectorizer() 
vectorizer.fit(doc) 
vector = vectorizer.fit_transform(doc) 
Fre_word = pd.DataFrame(data=vector.toarray(),
                        columns=vectorizer.get_feature_names()).T.reset_index()
Fre_word.columns = ['index','Freq']
Fre_word = Fre_word.sort_values('Freq',ascending=False).head(50)
Fre_word
```

```python
plt.figure(figsize=(24, 12))
d = {}
for k,v in zip(Fre_word['index'], Fre_word['Freq']):
    d[k] = int(v)
wordcloud = WordCloud(font_path='C:/Windows/Fonts/Microsoft YaHei/msyh.ttc',
                      width=1600,
                      height=800,
                      background_color="white").generate_from_frequencies(d)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
# plt.imshow(wordcloud, interpolation="bilinear")
```

```python
majorcate.head()
```

```python
# 前 50大科系
top50_majors = majorcate['major'].value_counts().reset_index().head(50)
top50_majors.columns = ['major', 'cnt']
top50_majors
```

```python
# 前50大科系的職缺
top50_jobdesc = pd.merge(left = majorcate,
                           right = top50_majors.loc[:,'major'],
                           how = 'inner',
                           on = 'major')
top50_jobdesc
```

### 薪資

```python
sal_major = pd.merge(left = top50_jobdesc,
                     right = df.loc[:,['index','nsalaryMin']],
                     how='left',
                     on = 'index')
sal_major = sal_major.groupby(['major','nsalaryMin']).size().reset_index()
sal_major.columns = ['major','nsalaryMin', 'cnt']
sal_major = sal_major.pivot_table(values='cnt',columns='nsalaryMin',index='major')
sal_major
```

### 職務

```python
jobcate_major = pd.merge(left = top50_jobdesc,
                         right = jobcate.loc[:,['index','desc']],
                         how='left',
                         on = 'index')
jobcate_major = jobcate_major.groupby(['major','desc']).size().reset_index()
jobcate_major.columns = ['major','desc', 'cnt']
jobcate_major = jobcate_major.sort_values(['major','cnt'],ascending=False)#.groupby('major').head(3)
jobcate_major
```

```python
jobcate_major
```
